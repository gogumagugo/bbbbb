```python
import pandas as pd


# 3번째 열을 선택하고 첫 번째 행의 '방문수' 제거
column_index = 0  # 3번째 열은 인덱스 0임을 확인

# '방문수' 문자열 제거
result_df.iloc[:, column_index] = result_df.iloc[:, column_index].replace({'방문수': None})

# 문자열에서 숫자만 추출하여 변환하는 함수
def convert_to_float(value):
    if isinstance(value, str):
        # 'M'이나 ','를 제거하고 숫자로 변환
        value = value.replace('M', '').replace(',', '')
        try:
            return float(value) * 1e6 if 'M' in value else float(value)
        except ValueError:
            return None
    return value

# 데이터 변환 및 NaN 값 처리
result_df.iloc[:, column_index] = result_df.iloc[:, column_index].apply(convert_to_float)

# 2행부터 100행까지의 데이터 선택
df_sub = result_df.iloc[1:100, column_index].copy()

# NaN 값 제거
df_sub = df_sub.dropna()

# 데이터 총합 계산
total_sum = df_sub.sum()

# 총합 출력
print("2행부터 100행까지의 총합:", total_sum)
```
